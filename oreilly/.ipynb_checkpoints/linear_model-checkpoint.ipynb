```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import os
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn import linear_model
```

```python
# read dataset
DATA_PATH = os.path.join('datasets/housing','housing.csv')
df = pd.read_csv(DATA_PATH)
df.head()
```

```python
# split dataset
train, test = train_test_split(df, test_size = 0.15)
```

```python
# build model
model = linear_model.LinearRegression()
```

```python
# sns.barplot(x = "median_income", y = "median_house_value", data = train)
train.plot(kind='scatter', x="median_income", y="median_house_value", alpha = 0.1) # alpha: transparency
```

```python
# Prepare the data
X_median_income = np.c_[train["median_income"]]
Y_median_house_value = np.c_[train["median_house_value"]]
Y_median_house_value
```

```python
# # Train the model
model.fit(X_median_income, Y_median_house_value)
```

```python
# # Make a prediction for Cyprus
X_new = [[4]]  
print(model.predict(X_new))
```

```python
# test
X_median_income_test = np.c_[test["median_income"]]
Y_median_house_value_test = np.c_[test["median_house_value"]]
Y_median_house_value_test 

Y_median_house_value_pred = model.predict(X_median_income_test) 
# Y_median_house_value_pred # list of list
# Accuracy is a classification metric. You can't use it with a regression. See the documentation for info on the various metrics.
# score = accuracy_score(Y_median_house_value_test, Y_median_house_value_pred)
# score

# MSE mean squared error
MSE = mean_squared_error(Y_median_house_value_test, Y_median_house_value_pred)
MSE

# r square
```

```python
# r squre
r2 = r2_score(Y_median_house_value_test, Y_median_house_value_test)
r2
```
